```python
import zipfile
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
zip_path = "archive (8).zip"
extract_to = "project_eleven_labs"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_to)
    print(f" Extracted files to: {extract_to}")
```

```python
extracted_files = os.listdir(extract_to)
print("Files in folder:", extracted_files)
```

```python
csv_file = next((file for file in extracted_files if file.endswith(".csv")), None)

if csv_file:
    csv_path = os.path.join(extract_to, csv_file)
    try:
        df = pd.read_csv(csv_path)
        print(" CSV Loaded Successfully!")
        display(df.head())
    except Exception as e:
        print(" Error reading CSV:", e)
else:
    print(" No CSV file found in the folder.")
```

```python
print(df.describe())
print("Median values:\n", df.median(numeric_only=True))
```

```python
# Histogram
sns.histplot(df['Age'].dropna(), kde=True)
plt.title("Age Distribution")
plt.show()

# Boxplot
sns.boxplot(y='Age', data=df)
plt.title("Boxplot of Age")
plt.show()
```

```python
sns.histplot(df['Fare'], kde=True)
plt.title("Fare Distribution")
plt.show()

sns.boxplot(y='Fare', data=df)
plt.title("Boxplot of Fare")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare']].dropna(), hue='Survived')
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival Count by Sex")
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival Count by Sex")
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

```python
sns.histplot(data=df, x='Age', hue='Survived', multiple='stack', bins=30)
plt.title("Age Distribution by Survival")
plt.show()
```
